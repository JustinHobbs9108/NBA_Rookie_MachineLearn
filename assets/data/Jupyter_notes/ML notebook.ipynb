```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv('Nba_injuries_small.csv')
df.head()
```

```python
pos_labels = {}
for idx, row in df.iterrows():
    if row["Position"] not in pos_labels:
        pos_labels[row["Position"]] = len(pos_labels)
print(pos_labels)
```

```python
for pos in pos_labels:
    df.replace(pos, pos_labels[pos], inplace=True)
df.head()
```

```python
data = df.values
X = data[:, 0:2]
y = data[:, 2:6]
```

```python
print(X)
```

```python
spraintarget = df["Strains/Sprains_y_True"]
spraintarget_names = ["no", "yes"]
```

```python
spraindata = df.drop(["Strains/Sprains_y_True", "Dislocations_y_True", "Arm Fractures_y_True", "Leg Fractures_y_True"], axis=1)
feature_names = spraindata.columns
spraindata.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, spraintarget, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
sprainrf = RandomForestClassifier(n_estimators=200)
sprainrf = sprainrf.fit(X_train, y_train)
sprainrf.score(X_test, y_test)
```

```python
sorted(zip(sprainrf.feature_importances_, feature_names), reverse=True)
```
