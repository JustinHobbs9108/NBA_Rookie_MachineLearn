```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
nba_hurt = pd.read_csv('nba_hurtsKW2.csv')
nba_hurt
nba_hurt = nba_hurt.drop([0], axis = 0)
nba_hurt.dropna(how = "any")
```

```python
nodups_hurt=nba_hurt.drop_duplicates(subset=['Date_x','name','Notes_x'],keep='first') 
fixed=nodups_hurt.drop(columns=['Estimated Medical Costs (2012)_y','year_start_x','year_end_x','Date_x','birth_date_x','Unnamed: 0_x','Unnamed: 0.1_x'])
fixed.rename(columns={'position_x':'Position'},inplace=True)
fixed.head()
```

```python
fixed.groupby(['name'])['name'].count()
```

```python
all_nba_players=pd.read_csv('nba_all_players_2019_roster.csv')
```

```python
nba_players=all_nba_players.drop(columns=['Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13','Unnamed: 14'])
nba_players=nba_players.drop(columns=['Jersey Number'])
nba_players['name']=nba_players['First Name']+ ' ' + nba_players['Last Name']
nba_players.drop(columns=['First Name','Last Name'])
nba_players=nba_players[['Team Name','name','Position','Height (in)','Weight (lbs)']]
nba_players=nba_players.replace('Guard','G')
nba_players=nba_players.replace('Center','C')
nba_players=nba_players.replace('Forward','F')
nba_players=nba_players.replace('Center-Forward','F-C')
nba_players=nba_players.replace('Forward-Center','F-C')
nba_players=nba_players.replace('Forward-Guard','F-G')
nba_players=nba_players.replace('Guard-Forward','F-G')
nba_players
```

```python
class Injury:
    def __init__(self, player, notes, date):
        self.player = player
        self.notes = notes
        self.date = date
    
    def __eq__(self, other):
        return self.player == other.player and self.notes == other.notes and self.date == other.date
    
    def __repr__(self):
        return f"Injury(Player: {self.player}, Notes: {self.notes}, Date: {self.date})"
```

```python
test = Injury("Jeff", "Hurty", "2019-08-27")
```

```python
unique_injury_list = []
for index,rows in nba_hurt.iterrows():
    INJ=Injury(rows['name'],rows['Notes_x'],rows['Date_x'])
    if INJ not in unique_injury_list:
        unique_injury_list.append(INJ)
```

```python
fdummies = pd.get_dummies(data = fixed, columns = ["Strains/Sprains_y", "Dislocations_y","Arm Fractures_y","Leg Fractures_y"])
fdummies = fdummies[['name','Strains/Sprains_y_True','Dislocations_y_True','Arm Fractures_y_True','Leg Fractures_y_True']]
fdummies=fdummies.groupby(['name']).sum()
```

```python
nba_players=nba_players.merge(fdummies, on='name',how='left')
nba_players
```
